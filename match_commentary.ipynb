```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.espncricinfo.com/series/india-in-sri-lanka-2024-1442984/sri-lanka-vs-india-3rd-odi-1442992/ball-by-ball-commentary'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
class_name = "lg:hover:ds-bg-ui-fill-translucent ds-hover-parent ds-relative"
```

```python
divs = soup.find_all('div', class_=class_name)
divs.reverse()

divs[0]
```

```python
import pandas as pd

df = pd.DataFrame(columns = ["ball", "action", "commentary"])
df
```

```python
for div in divs:
    ball_div = div.find('div', class_="lg:ds-flex lg:ds-items-center lg:ds-px-2")
    ball_no = ball_div.find('span', class_="ds-text-tight-s ds-font-regular ds-mb-1 lg:ds-mb-0 lg:ds-mr-3 ds-block ds-text-center ds-text-typo-mid1").text.strip()

    result_div = div.find('div', class_="xl:ds-w-[730px]")
    p_div = result_div.find('div', class_="ds-ml-4 lg:ds-ml-3 ds-text-typo-mid1")
    action_div = p_div.find('div', class_="ds-leading-none ds-mb-0.5")
    action = action_div.text.strip()

    commentary = result_div.find('div', class_="first-letter:ds-capitalize").text.strip()
    #print(commentary)

    row = [ball_no, action, commentary]
    length = len(df)
    df.loc[length] = row

df
```

```python
df.to_csv('match_commentary.csv', index=False)
```
